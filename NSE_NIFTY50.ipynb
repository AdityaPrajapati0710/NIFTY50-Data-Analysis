```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Perceptron model
class Perceptron:
    def __init__(self, learning_rate=0.1):
        self.weights = np.random.rand(2)
        self.learning_rate = learning_rate

    def predict(self, inputs):
        summation = np.dot(inputs, self.weights)
        return 1 if summation > 0 else 0

    def train(self, training_inputs, labels):
        for inputs, label in zip(training_inputs, labels):
            prediction = self.predict(inputs)
            self.weights += self.learning_rate * (label - prediction) * inputs

# Training data
training_inputs = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
labels = np.array([0, 0, 0, 1])  # Example logic: AND gate

# Create Perceptron
perceptron = Perceptron()

# Animation function
def animate(i):
    perceptron.train(training_inputs, labels)
    plt.cla()
    for point, label in zip(training_inputs, labels):
        color = 'r' if label == 0 else 'g'
        plt.scatter(point[0], point[1], c=color)
    x = np.linspace(-1, 2, 10)
    if perceptron.weights[1] == 0:
        y = np.zeros_like(x)
    else:
        y = -(perceptron.weights[0] * x) / perceptron.weights[1]
    plt.plot(x, y, '-b')
    plt.xlim(-0.5, 1.5)
    plt.ylim(-0.5, 1.5)
    plt.title("Perceptron Training")
    plt.xlabel("Input 1")
    plt.ylabel("Input 2")

# Create animation
fig, _ = plt.subplots()
ani = FuncAnimation(fig, animate, frames=20, interval=1000)

# Display the animation
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Perceptron model
class Perceptron:
    def __init__(self, learning_rate=0.1):
        self.weights = np.random.rand(2)
        self.learning_rate = learning_rate

    def predict(self, inputs):
        summation = np.dot(inputs, self.weights)
        return 1 if summation > 0 else 0

    def train(self, training_inputs, labels):
        for inputs, label in zip(training_inputs, labels):
            prediction = self.predict(inputs)
            self.weights += self.learning_rate * (label - prediction) * inputs

# Training data
training_inputs = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
labels = np.array([0, 0, 0, 1])  # Example logic: AND gate

# Create and train Perceptron
perceptron = Perceptron()
for _ in range(20):
    perceptron.train(training_inputs, labels)

# Plotting
plt.figure()
for point, label in zip(training_inputs, labels):
    color = 'r' if label == 0 else 'g'
    plt.scatter(point[0], point[1], c=color)

x = np.linspace(-1, 2, 10)
if perceptron.weights[1] == 0:
    y = np.zeros_like(x)
else:
    y = -(perceptron.weights[0] * x) / perceptron.weights[1]
plt.plot(x, y, '-b')
plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)
plt.title("Perceptron Decision Boundary")
plt.xlabel("Input 1")
plt.ylabel("Input 2")
plt.show()
```
